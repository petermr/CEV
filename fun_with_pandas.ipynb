# Reading csv files

Through pandas you can read several types of file like csv,xls,hsv etc. More about this can be found [here](https://) 

Our concerned file type is CSV therefore we have to do read_csv. More information about read_csv can be found [here](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html)
There are lot of attributes for read_csv. Below I have mention few important once:
Header: 

1.   **header**:  If set to None, there will be no header for your dataframe. DEfault is 'infer' which autometically detects the header from csv
2.   **index** :  Similar to header if set to None it wont return any indexes of you dataframe.
3.   **names**  : List of column names to use. If the file contains a header row, then you should explicitly pass header=0 to override the column names. Duplicates in this list are not allowed.

# Combining two csv
When you have several csv and want to combine them, Pandas concat function comes really handy. 
In the code below we are appending two csv by first converting them to dataframes and then concatenating them with help of concat function.

You can find more about **pd.concat** [here](https://)


```python
#appending two csv files :

df1 = pd.read_csv('/content/Terpene and Other Plant Pathway Figures.csv')
df2 = pd.read_csv('/content/Terpene and Other Plant Pathway Figures (3).csv')
df_new = pd.concat([df1,df2], axis = 0).reset_index() ##horizontly
#df_new = pd.concat([df1,df2], axis = 1) ##vertically
```
